```python
import pandas as pd
import numpy as np
import pickle
```

```python
data = pd.read_csv('data/table_pvp.csv', sep=';', encoding='utf-8')
```

```python
data.head(3)
```

```python
d = {x: k for k, x in enumerate(sorted(set(data['strategy_1'])))}
with open('dicts/strategy', mode='wb') as f:
    pickle.dump(d, f)
data['strategy_1'] = data['strategy_1'].apply(lambda x: d[x])
data['strategy_2'] = data['strategy_2'].apply(lambda x: d[x])
```

```python
d = {x: k for k, x in enumerate(sorted(set(data['pass_1'])))}
with open('dicts/pass', mode='wb') as f:
    pickle.dump(d, f)
data['pass_1'] = data['pass_1'].apply(lambda x: d[x])
data['pass_2'] = data['pass_2'].apply(lambda x: d[x])
```

```python
d = {x: k for k, x in enumerate(sorted(set(data['formation_1'])))}
with open('dicts/formation', mode='wb') as f:
    pickle.dump(d, f)
data['formation_1'] = data['formation_1'].apply(lambda x: d[x])
data['formation_2'] = data['formation_2'].apply(lambda x: d[x])
```

```python
data.head(3)
```

```python
data.to_csv('data/data_normal.csv', index=False, sep=';', encoding='utf-8')
```

## Мешок слов

```python
X_const = data.values[:, :5]
X_cat = data.values[:, 5:]
```

```python
from sklearn.preprocessing import OneHotEncoder
X_cat = OneHotEncoder().fit_transform(X_cat)
X_cat = X_cat.todense()
```

```python
X_ = np.hstack((X_const,X_cat))
```

```python
df = pd.DataFrame(X_)
```

## Идеи

```python
data['power_1'] = data['power_1'].apply(lambda x: int(round(x)))
data['power_2'] = data['power_2'].apply(lambda x: int(round(x)))
```

```python
data['pow'] = data['power_1'] - data['power_2']
data['powA'] = (data['power_1'] + data['power_2']) / 2
del data['power_1']
del data['power_2']
```

```python
data['tac'] = data['tactics_1'] - data['tactics_2']
data['tacA'] = (data['tactics_1'] + data['tactics_2']) / 2
del data['tactics_1']
del data['tactics_2']
```
